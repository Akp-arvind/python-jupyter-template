### Exercise 4: Pie Chart with Percentages

```python
# TASK: Create a pie chart with the following data: labels = ['Python', 'Java', 'C++', 'JavaScript'] and sizes = [40, 25, 20, 15].
# Display the percentages on the chart using autopct.


import matplotlib.pyplot as plt

# Step 1: Define Data
labels = ['Python', 'Java', 'C++', 'JavaScript']
sizes = [40, 25, 20, 15]
colors = ['blue', 'red', 'green', 'orange']  # Assign different colors
explode = (0.1, 0, 0, 0)  # Slightly separate Python slice for emphasis

# Step 2: Create Pie Chart
plt.figure(figsize=(7, 7))  # Set figure size
plt.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%', startangle=140, explode=explode, shadow=True)

# Step 3: Customize Plot
plt.title("Programming Language Usage Distribution", fontsize=14)  # Add title

# Step 4: Display the Pie Chart
plt.show()
```
