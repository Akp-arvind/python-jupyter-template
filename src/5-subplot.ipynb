### Exercise 5: Subplot Layout

```python
# TASK: Create a 2x2 subplot layout.
# Plot a line chart in the first subplot, a bar chart in the second, a scatter plot in the third, and a pie chart in the fourth.


import matplotlib.pyplot as plt
import numpy as np

# Create a figure with a 2x2 grid of subplots
fig, axs = plt.subplots(2, 2, figsize=(10, 8))  # 2 rows, 2 columns

# 1st Subplot: Line Chart
x = np.linspace(0, 10, 100)  # Generate 100 points from 0 to 10
y = np.sin(x)  # Sine function
axs[0, 0].plot(x, y, color='blue', linewidth=2)
axs[0, 0].set_title("Line Chart: Sine Wave")
axs[0, 0].set_xlabel("X")
axs[0, 0].set_ylabel("sin(X)")
axs[0, 0].grid(True, linestyle='--', alpha=0.7)

# 2nd Subplot: Bar Chart
categories = ['A', 'B', 'C', 'D']
values = [5, 7, 3, 9]
colors = ['red', 'blue', 'green', 'purple']
axs[0, 1].bar(categories, values, color=colors)
axs[0, 1].set_title("Bar Chart")
axs[0, 1].set_xlabel("Categories")
axs[0, 1].set_ylabel("Values")
axs[0, 1].grid(axis='y', linestyle='--', alpha=0.7)

# 3rd Subplot: Scatter Plot
x_scatter = [1, 2, 3, 4, 5]
y_scatter = [2, 4, 6, 8, 10]
axs[1, 0].scatter(x_scatter, y_scatter, color='green', marker='o', s=100)
axs[1, 0].set_title("Scatter Plot")
axs[1, 0].set_xlabel("X Values")
axs[1, 0].set_ylabel("Y Values")
axs[1, 0].grid(True, linestyle='--', alpha=0.7)

# 4th Subplot: Pie Chart
labels = ['Python', 'Java', 'C++', 'JavaScript']
sizes = [40, 25, 20, 15]
axs[1, 1].pie(sizes, labels=labels, autopct='%1.1f%%', startangle=140, colors=['blue', 'red', 'green', 'orange'])
axs[1, 1].set_title("Pie Chart")

# Adjust layout for better spacing
plt.tight_layout()

# Show the plots
plt.show()
```
