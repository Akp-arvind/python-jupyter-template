### Exercise 1: Line Plot Customization

```python
# TASK: Create a line plot with x values ranging from 0 to 10 and y values as the square of x.
# Customize the plot by adding a title, labels for both axes, and a grid.

import numpy as np
import matplotlib.pyplot as plt

# Step 1: Define x values (0 to 10)
x = np.linspace(0, 10, 100)  # 100 points for smooth curve
y = x**2  # y values as the square of x

# Step 2: Create the plot
plt.figure(figsize=(8, 5))  # Set figure size
plt.plot(x, y, color='b', linestyle='-', linewidth=2, label=r'$y = x^2$')  # Line plot

# Step 3: Customize the plot
plt.title("Line Plot of y = xÂ²", fontsize=14)  # Title
plt.xlabel("X Values", fontsize=12)  # X-axis label
plt.ylabel("Y Values (Square of X)", fontsize=12)  # Y-axis label
plt.grid(True, linestyle='--', alpha=0.7)  # Add grid with dashed lines
plt.legend()  # Show legend

# Step 4: Display the plot
plt.show()
```
