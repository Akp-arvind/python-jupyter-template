### Exercise 3: Scatter Plot with Annotations

```python
# TASK: Create a scatter plot with x = [1, 2, 3, 4, 5] and y = [2, 4, 6, 8, 10].
# Annotate each point with its (x, y) value, and set the title as 'Scatter Plot with Annotations'.


import matplotlib.pyplot as plt

# Step 1: Define Data
x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]

# Step 2: Create Scatter Plot
plt.figure(figsize=(7, 5))  # Set figure size
plt.scatter(x, y, color='blue', marker='o', s=100, label="Data Points")  # Scatter plot with blue dots

# Step 3: Annotate Each Point
for i in range(len(x)):
    plt.text(x[i], y[i], f"({x[i]}, {y[i]})", fontsize=11, ha='right', va='bottom', color='black')

# Step 4: Customize Plot
plt.title("Scatter Plot with Annotations", fontsize=14)  # Title
plt.xlabel("X Values", fontsize=12)  # X-axis label
plt.ylabel("Y Values", fontsize=12)  # Y-axis label
plt.grid(True, linestyle='--', alpha=0.7)  # Add grid lines
plt.legend()  # Show legend

# Step 5: Display the Plot
plt.show()
```
