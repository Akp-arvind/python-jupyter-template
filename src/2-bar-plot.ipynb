### Exercise 2: Bar Plot with Colors

```python
# TASK: Create a bar plot with the following data: categories = ['A', 'B', 'C', 'D'] and values = [5, 7, 3, 9].
# Use different colors for each bar and add a title to the plot.


import matplotlib.pyplot as plt

# Step 1: Define data
categories = ['A', 'B', 'C', 'D']
values = [5, 7, 3, 9]
colors = ['red', 'blue', 'green', 'purple']  # Different colors for each bar

# Step 2: Create the bar plot
plt.figure(figsize=(8, 5))  # Set figure size
plt.bar(categories, values, color=colors)  # Create bar chart with custom colors

# Step 3: Customize the plot
plt.title("Bar Plot of Categories", fontsize=14)  # Add title
plt.xlabel("Categories", fontsize=12)  # X-axis label
plt.ylabel("Values", fontsize=12)  # Y-axis label
plt.grid(axis='y', linestyle='--', alpha=0.7)  # Add horizontal grid lines
plt.xticks(fontsize=11)  # Customize x-axis tick labels
plt.yticks(fontsize=11)  # Customize y-axis tick labels

# Step 4: Display the plot
plt.show()
```
